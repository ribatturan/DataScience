```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('audit_risk_original.csv')
```

```python
corr_mat = df.corr()
fig, ax = plt.subplots(figsize=(12, 12))
sns.heatmap(corr_mat, annot=True, ax=ax,
            fmt='.1g', cmap='coolwarm', linewidths=.5,
            vmin=-0.6, vmax=1, center= 0)
plt.show()
```

```python
df.info()
```

```python
deleted_col = ['LOCATION_ID','TOTAL']
df.drop(deleted_col, axis = 1, inplace = True)
```

```python
df['Money_Value'].fillna((df['Money_Value'].mean()), inplace = True)
```

```python
df.isna().sum()
```

```python
X = df.drop(['Risk'],axis = 1)
```

```python
y = df['Risk']
```

```python
model = DecisionTreeClassifier(criterion = 'gini', max_depth = 5)
```

```python
metrics = ['average_precision','accuracy','f1','f1_weighted','precision','roc_auc']
for metric in metrics:
    score = cross_val_score(model,X,y,cv = 5,scoring = metric)
    print("Model : {}, Metric : {}, Score : {}".format(model,metric,score))
```

```python
from sklearn import metrics, model_selection,datasets,svm
model.fit(X,y)
metrics.plot_precision_recall_curve(model,X,y)
plt.show()
```
